```python
import pandas as pd
import numpy as np

cols=["v1", "v2"]
messages=pd.read_csv("spam.csv", usecols=cols, encoding="latin-1")
print(messages)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

messages["v1"].replace('ham', '0', inplace=True)
messages["v1"].replace('spam', '1', inplace=True)
messages.drop(1, inplace=True)
print(messages)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
messages["v2"]=messages["v2"].astype("str")
le.fit_transform(messages["v2"])

x=messages.iloc[:, 0:1]
y=messages["v1"]
print(x.shape)
print(y.shape)
y=y.transpose()

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.9)


from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver="saga")

lr.fit(x, y)

pred=lr.predict(x_test)
print(pred)

from sklearn.metrics import accuracy_score
model_accuracy=round(lr.score(x, y)*100, 2)
print(f"Model Accuracy :{ model_accuracy}")

model_accuracy=round(lr.score(x_test, y_test)*100, 2)
print(f"Model Accuracy :{ model_accuracy}")

print(lr.predict(x_train))

import sklearn
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(max_iter=500, solver="lbfgs", hidden_layer_sizes=(2,2,2), activation="relu")
mlp.fit(x, y)
pred2=mlp.predict(x_test)
print(pred2)
pred2=mlp.predict(x_train)
print(pred2)
model_accuracy=round(mlp.score(x_test, y_test)*100, 2)
print(f"Model Accuracy :{ model_accuracy}")
```
